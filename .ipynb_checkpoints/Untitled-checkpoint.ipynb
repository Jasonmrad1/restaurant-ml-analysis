```python
import pandas as pd

# Load dataset
df = pd.read_csv("restaurant_data.csv")

# Basic info
print(df.shape)        # rows, columns
print(df.isnull().sum())  # no missing values
print(df.describe(include="all"))  # numeric + categorical summary
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(8,5))
sns.histplot(df["Revenue"], bins=30, kde=True)
plt.title("Distribution of Restaurant Revenue")
plt.show()
```

```python
target = "Revenue"

# 1 Correlation with Revenue for numeric columns
numeric_cols = df.select_dtypes(include=["int64", "float64"]).columns.tolist()
numeric_cols.remove(target)  # remove target itself

print("Correlation with Revenue:")
for col in numeric_cols:
    corr = df[col].corr(df[target])
    print(f"{col}: {corr:.3f}")

# 2 Visualization
for col in df.columns:
    if col == target:
        continue
    
    plt.figure(figsize=(7,5))
    
    if df[col].dtype in ["int64", "float64"]:  # Numeric feature
        sns.scatterplot(x=df[col], y=df[target])
        plt.title(f"{col} vs {target}")
        plt.xlabel(col)
        plt.ylabel(target)
        
        # Optional: regression line
        sns.regplot(x=df[col], y=df[target], scatter=False, color='red')
        
    else:  # Categorical feature
        sns.boxplot(x=df[col], y=df[target])
        plt.title(f"{col} vs {target}")
        plt.xlabel(col)
        plt.ylabel(target)
    
    plt.tight_layout()
    plt.show()
```
